```python
import matplotlib.pyplot as plt
```

```python
exp_arr = np.random.exponential(size=(100000,))
uniform_arr = np.random.rand(100000)
normal_arr = np.random.randn(100000)

plt.hist(exp_arr, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_arr, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normal_arr, density=True, bins=100, histtype="step", color="blue", label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

```
